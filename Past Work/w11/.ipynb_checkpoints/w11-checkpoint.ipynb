<p style="border-radius:3px;background-color:steelblue;padding:5px;color:white;">The main purpose of EDA is to ... </p>
<ol>
    <li>get to know our data before making assumptions,</li>
    <li>before finalizing a qualified researchable question, </li>
    <li>identify &ldquo;errors&rdquo; in the data themselves</li>
    <li>appropriateness or applicability of the data to the questions</li>
    <li>range and domain of the data</li>
    <li>and ultimately allow us to extract groups for statistical testing to identify &ldquo;interesting events&rdquo;</li>
    <li>potentially interesting data sub-sets are subjected to further statistical tests and </li>
    <li>use plots to explore the data as well as later present, describe, explain, and explore questions in a visual language</li>
    <li>interpreted by the data scientist and a domain expert</li>
    <li>for insights into the data - new knowledge - with the correct statistical and domain background to warrant our interpretations.</li>
</ol>

```python
# NOTE: typically import libraries for data
import numpy as np
import pandas as pd

# and libraries for visualizing/plotting the data
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# load the dataset
# remember, hard-coding addresses isn't recommended in practice.
# It's done here 'cause this beta OS won't find the file otherwise (or operator error!)
#books = pd.read_csv("data/classics.csv")
books = pd.read_csv("data/classics.csv")

books.head()
```

<p style="border-radius:3px;background-color:steelblue;padding:5px;color:white;">Having gotten to know the file of data, let&rsquo;s look at some facets of the data themselves.</p>
<p style="font-size:18px; font-family:Palatino, serif;">Note the language codes ('en', 'de', 'es', etc.)  Whenever possible use professional and industry standards in your data.  These codes refer to 
    <a href="https://www.iso.org/iso-639-language-codes.html" target="new">ISO639 Org homepage</a> and more info at <a href="https://iso639-3.sil.org/code_tables/639/data" target="new">SIL</a>.  This standard is extremely common and often required in networked data and systems.</p>

```python
print(books['bibliography.languages'].value_counts())
print(books['metrics.statistics.words'].value_counts())
print(books['metrics.statistics.words'].mode())
print(books['metrics.statistics.words'].value_counts().idxmax())
```

```python
books.describe()
```

<hr />
<p><b>Starting to clean the data and our research questions</b> - Nones and NaN and useless fields. Our data might not confirm to the needs of our research question nor to the requirements of a statistical test. When writing up research reports, be sure to identify the original state of the data and why you dropped or altered any data and confirm the reduced set is robust enough for statistical test(s) you choose.</p>
<p>A great example of data, the statistical tests, and how to report them in APA is in Hair, et al. <i>Multivariate analysis.</i></p>

```python
print(books.isnull().sum()) # compare to using boolean values
```

<hr />
<p style="border-radius:3px;background-color:steelblue;padding:5px;color:white;">Summary and notes</p>
<p> ... notice  that when working with data some features aren&rsquo;t within range or are missing too many values, or there are other reasons to exclude from the next step of exploration and analysis.</p>
<p>Notice in the below box the chaining of methods - you may need to force the order of operations with ( ) if you receive errors and weird results.<p>

```python
books_final = books.isnull().sum().sort_values(ascending=False)
percent = (books.isnull().sum()/books.isnull().count()).sort_values(ascending=False)
books_missing = pd.concat([books_final,percent], axis=1, keys=['Total','Percent'])
books_missing.head(20)

# notice all the 0 values ... can these be used in your research?  Deleted? 
```

```python
print(books['bibliography.publication.full'])
```

```python
books['bibliography.publication.full'] = pd.to_datetime(books['bibliography.publication.full'], errors = 'coerce')
books['month'] = books['bibliography.publication.full'].dt.month
books.info()
```

<p style="border-radius:3px;background-color:steelblue;padding:5px;color:white;">Example 2: Airline Data</p>
<p>Scenario: We&rsquo;re given this dataset ... and so need to get to know this file &amp; the data ... we see Atlanta is often impacted by delays.  We think weather and airline delays are related, so let&rsquo;s explore ... Coming shortly we&rsquo;ll do much more with <code>groupby</code> and aggregate to extract data for testing to see what else we can learn.</p>

```python
#https://corgis-edu.github.io/corgis/csv/airlines/
import pandas as pd
df = pd.read_csv('data/airlines.csv')

# df.to_string() # if we want to see all the header and text data.

#1 head() and tail()
df.head()
```

```python
df.tail()
```

```python
df.info()
# note that the Non-Null Count shows the # of null entries - there are none 'cause
# we have 4408 rows and all show as having data (non-null)
```

<hr />
<h2 style="background-color: #003262; color:white; border-radius: 4px; padding: 8px;">Check for data having empty cells, data in the wrong format, out-of-range data, duplicates.</h2>
<p>In the following cells, imagine we have an Airline dataset to explore ... working from getting to know the data to asking a few questions of them. We notice a lot of delays in Atlanta, so let's check it out.</p>
<p>We have fundamental ideas about what to do with a dataset for preparing it for exploration &amp; analysis ... but at his point what to look for and how to find it depends on your project. No training wheels... </p>

```python
df.dropna(inplace=True)
df2 = df[["Airport.Code","Time.Label","Time.Month","Time.Year","Statistics.Carriers.Total","Statistics.# of Delays.Carrier", "Statistics.Minutes Delayed.Weather"]] 
df2.head()
```

```python
# Those column titles are too long ... let's rename them to our purpose. 
df2.columns = ['Airport','Date','Month','Year','Total Flights','Delays','Weather']
df2.head(10)
```

```python
# get a sense of what's to be expected - what's the average delay of this data set (over ? years)
print("For the period", df2['Year'].min()," to ", df2['Year'].max())
print("The average national delay in minutes was ", end='')
print(df2['Delays'].mean())
```

```python
df2.sort_values(by='Delays', ascending=False)
```

```python
df2.loc[df2['Airport'] == 'ATL'], ['Delays']
```

```python
df2.groupby('Airport', as_index=False)['Delays'].mean().sort_values(by='Delays', ascending=False)

# you can achieve the same output with 
# df2.groupby('Airport').Delays.mean()
```

```python
#NOTE! there are at least 5 ways to do this same output ... 
def minToHours(t):
    return t/60 % 60

ATL_Delays  = df2[df2['Airport']=='ATL']['Delays'].mean()
ATL_Delays

print(minToHours(ATL_Delays))
```

```python
df2.sort_values(by = ['Airport','Weather']).groupby('Airport')
df2
```

```python
# is there a relationship between weather and delays?
df3 = df2[['Weather','Delays']]
df3.corr()

# .7 is a pretty strong positive relationship!  
# (leaves us about .3 of the error to explain)
```

```python
#display(df2.query('Airport = ATL & Year >= 2016,values'))
display(df2[(df2['Airport'] == "ATL") & (df2['Year'] == 2016)][['Airport','Year', 'Total Flights','Delays']])
```
