<p style="font-family:Baskerville; color:red;font-size:22px;"><i>A Demonstration Project used this notebook:</i></p>
<a href="images/rhode-island-county-map.png" target="new"><img style="float:left; width:220px;" src="images/rhode-island-county-map.png"></a><p>Say you&rsquo;re faced with a decision and want to use data to help select your new home.  Where to live in Rhode Island?  Explore data about the state from official records to get to know the area of our study and then ask questions that allow us to make decisions.</p>
<p>Notice there are a couple of levels of granularity: the entire state, the county, and the individual cities (all five of them) and the towns.</p>
<img style="float:right; width:150px;" src="images/goatisland.png"/><br />
<p style="font-size:9px; font-family:'Avenir Next', sans-serif;">The map shows the towns and cities of the Ocean State (including prof. Beno√Æt&rsquo;s hometown of Scituate); on the right is an image of Goat Island lighthouse, now a BnB you can rent on a small island in Newport Harbor. Recommended!</p>

```python
import pandas as pd

df = pd.read_csv('ri_data.txt')
df.head()
```

<hr /><p style="font-family:Baskerville; color:red;font-size:22px;"><i>Becoming familiar with the file</i></p>
<p>Given this small set, I want to group by county and explore...  How many sets will I find?  Who are the members thereof? </p>

```python
df_county = df.groupby('County')
print("Number of groups: ", df_county.ngroups)
```

```python
print("Groups: ", df_county.groups)
print("-"*40)
# tells us the # of towns in each county
print("Towns per County: ", df_county.size())
```

```python
df_county.first()
```

```python
df_newport = df_county.get_group('Newport')
df_newport
```

<hr /><p style="font-family:Baskerville; color:red;font-size:22px;"><i>Check out some basic stats...</i></p>

```python
# great.  We have our subset ... let's get some basic
# stats ...
df_newport.groupby('Town').Household.agg(['min','max','mean'])
```

<p>In the above case, it turns out own data already is mean/max/min... so let's just see the min/max</p>

```python
df.groupby('Town').Household.agg(
    town_max=('max'),town_min=('min'))
```

<hr /><p style="font-family:Baskerville; color:red;font-size:22px;"><i>Functions or Lambda expressions for conditions: a combination of the stringing of methods + lambda...</i></p>

```python
# we're looking for towns with household income
# greater than 80000
def town_income(x):
    i = x.mean()
    return True if i > 80000 else False

df_x = df.groupby('Town').Household.agg(['mean',town_income])
# this can be written with a lambda ... 
# replace 'town_income' with lambda x : 
# True if x.mean() > 80000 else False
```

```python
df_wealthy = df.groupby('Town').Household.agg(
    ['mean',lambda x : True if x.mean() > 80000 else False])
df_wealthy
```

```python
df_compare = df.loc[:, ['Town','Household', 'PerCap']]
df_compare.groupby(['Town','PerCap']).mean()
```

<hr />
<p style="font-family:Baskerville; color:red;font-size:22px;"><i>Zeroing in ...</i>.</p><p>After our initial review of the data we have a sense of what we might want to focus on ... here, say we want a single locale, Newport.</p>

```python
df_compare.loc[df_compare['Town'] == 'Newport']
```

<p>Scenario: you have 2+ datasets to merge and to explore because you want to <i>ask questions of the data</i>, such as what's the most educated and well-off <i>towns</i> (as our unit of analysis) of the state to move to? ... <p style="font-family:Baskerville; color:red;font-size:32px;line-height: 36px;"><i>Crosstabs</i></p>
<p>The pandas crosstab function builds a cross-tabulation table that can show the frequency with which certain groups of data appear. </p>
<p>Using a well-known dataset from UCI Machine Learning Repository [https://archive-beta.ics.uci.edu/]</p>
<p>For more about crosstabs versus pivot, see <a href="https://chenzheruc.github.io/tutorial/tutorial_crosstabpivot.htm" target="new">https://chenzheruc.github.io/tutorial/tutorial_crosstabpivot.htm</a></p>

```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_excel("sales-funnel.xlsx")
df.head()
```

<p style="font-family: Baskerville; color:red; font-size:18px;"><i>Definition</i></p>
<blockquote>
    <p>A contingency table is a data aggregating only by counting item.
People (and pandas) use the term "pivot" (formerly a MS-copyrighted term!)</p>
<p>A pivot table is a table of grouped values that aggregates the individual items of a more extensive table (such as from a database, spreadsheet, or business intelligence program) within one or more discrete categories. This summary might include sums, averages, or other statistics, which the pivot table groups together using a chosen aggregation function applied to the grouped values.</p>
    <p>The choice of field(s) to group depend on our task - what fields and groups interest us.</p>
    </blockquote>

```python
pd.pivot_table(df,index=["Name","Rep","Manager"])
```

<hr/>
<p>Imagine the above data table held a lot more!  It'd be useless to try to find any datum of significance; Grouping data by some larger concept, like job title, city/town, patient group, etc., make it easier for us to <i>use</i> the data.</p>

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
```

```python
# DataFrames like SQL: query
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
```

```python
pd.crosstab(df.Rep, df.Price)
```

```python
pd.crosstab(df.Rep, df.Price, margins=True, margins_name="Total")
```

```python
sns.heatmap(pd.crosstab([df.Product, df.Price], [df.Rep, df.Quantity]),
            cmap="YlGnBu", annot=True, cbar=False)
```

<p style="font-family:Baskerville; color:red;font-size:32px;line-height: 36px;"><i>Files: Structured, Semi-Structured, Unstructured Files, and some new efficient, special-purpose ones</i></p>
<p>A brief note (for Wk 14) about files.  A common division is <u>structured</u> data and <u>unstructured</u>.  Some folk include <u>semi-structured</u>.
</p>
<ol><li>
    <b>Structured</b>: usually something from an application, such as SQL, where the application itself facilitates creating data tables, conforming to a standard, that's more efficient for storage and retrieval.</li>
    <li><b>Unstructured</b>: is just a text file - the stream of letters or characters in a flat file, usually casually called .txt files.  Here we might include non-relational (NoSQL) databases.</li>
    <li><b>Semi-structured</b>: are text files where the structure of the file and the data are all just text.  Most common examples are .json and sgml-derivatives (.html, .svg, .xml, etc.): the metadata is used to extract meaningful subsets.</li>
</ol>
<p>Semi-structured files typically conform to some industry-specific practice and descriptive standards: for instance, XBRL (Extensible Business REporting Language), VRA4 (Visual Resources Descriptors, 4th ed.), MathML (Math Modeling Language), phyloXML (phylogenetic networks in genomics) and endless others.</p>
<hr />
<p>Notice there are efficiencies that can be build into data files (similar to the compressions and techniques used in binary/media files)  (thanks, Ray) - such as Apache Parquet (Databricks) project. [<a href="https://www.databricks.com/glossary/what-is-parquet" target="new">Apache Parquet homepage</a>]</p>

<p style="font-family:Baskerville; color:red;font-size:32px;line-height: 36px;"><i>Time Series</i></p>
<p>Time series analyses typically reveal trends, irregularities, time-dependent cycles and seasonality.</p>
<p>NB: There are also models popular <u>depending on the work domain</u> e.g., ARIMA (Auto-Regressive Integrated Moving Average) [to predict errors]; Moving Average [helps identify random noise and plotting to smooth out noise], techniques in Data Cleansing such as MCAR (Missing completely at random), MAR (missing at random), closest-fit, imputation using association rules, hidden Markov models and others.</p>
<p>Optionally, please visit the Data Cleaning presentation []</p>
<p></p>
<p>Naturally you&rsquo;ll have to work with, format, and manipulate date/time data.</p>
<p>Check out <a href="https://www.machinelearningplus.com/time-series/time-series-analysis-python/" target="new"> for a nice demo</a>.</p>

```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

# NOTE: the above strptime takes (string, format)
# but has been deprecated. Use pd.to_datetime().
# see https://pandas.pydata.org/docs/reference/api/pandas.Timestamp.strptime.html

series = read_csv('shampoo.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser = parser)
print(series.head())
```

```python
series.plot()
pyplot.show()
```

```python
autocorrelation_plot(series)
pyplot.show()
```
