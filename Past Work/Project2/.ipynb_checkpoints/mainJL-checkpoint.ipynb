https://www.theguardian.com/society/2022/sep/07/hateful-tweets-multiply-in-extreme-temperatures-us-analysis-finds


A study analyzed 4 billion geo-located tweets in the US and found that hateful tweets increase significantly with extreme temperatures. The research, published in The Lancet Planetary Health, reported up to a 22% rise in racist, misogynistic, and homophobic tweets when temperatures exceeded 42°C and a 12% increase when temperatures fell below -3°C. The study, conducted by the Potsdam Institute for Climate Impact Research, used machine-learning algorithms to identify approximately 75 million hate tweets in 773 US cities between 2014 and 2020. The research suggests that people tend to exhibit more aggressive online behavior in temperature extremes, with the lowest number of abusive messages occurring at temperatures between 15-18°C. The study also noted that hate speech occurred across all climate zones, independent of factors such as income, political views, or religion. The impact of the climate crisis on hate speech could potentially be reduced by addressing corporate models that incentivize polarizing content, according to Diego Naranjo from the European Digital Rights network. The study highlights the serious threat of online hate speech to mental health, especially for young people and marginalized groups.

```python
grouped_df = hc_df.groupby('State')['Population'].sum().reset_index()

# Plot populations by type
grouped_df.plot(x='State', y='Population', kind='bar', figsize=(10, 6), color='blue')
plt.title('Populations of States')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

### Ohio has the most race motivated hate crimes

```python
ohio_df = hc_df[hc_df['State'] == 'Ohio']
```

```python
ohio_groups_df = ohio_df.groupby('Agency name')['Race'].sum().reset_index()

# Plot populations by type
ohio_groups_df.plot(x='Agency name', y='Race', kind='bar', figsize=(10, 6), color='blue')
plt.title('Race Crimes in Ohio Cities')
plt.xlabel('Cities')
plt.ylabel('Crimes')
plt.xticks(rotation=90)
plt.show()
```

## Race crimes in Columbus are clearly an outlier. Why? 

```python
print(ohio_df[ohio_df['Agency name'] == 'Columbus'])
```

#### Population of Columbus is quite high. We should create crimes committed per capita columns (We have 0 NaN values in all of the crime columns and the population columns, so no need to worry about removing rows)

```python
hc_df['Race Crime to Population Ratio'] = hc_df['Race'] / hc_df['Population']
```

#### Lets check Columbus again

```python
ohio_df[ohio_df['Agency name'] == 'Columbus']
```

#### Lets plot again, but with the per capita number

```python
ohio_groups_df = ohio_df.groupby('Agency name')['Race Crime to Population Ratio'].sum().reset_index()

# Plot populations by type
ohio_groups_df.plot(x='Agency name', y='Race Crime to Population Ratio', kind='bar', figsize=(10, 6))
plt.title('Race Crimes in Ohio Cities per Capita')
plt.xlabel('Cities')
plt.ylabel('Crimes')
plt.xticks(rotation=90)
plt.show()
```

## Columbus has a relatively average Race crime rate for its population. Separately, Richwood seems to have the highest rate. Why? 

```python
ohio_df[ohio_df['Agency name'] == 'Richwood']
```
